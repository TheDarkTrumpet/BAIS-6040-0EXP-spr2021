# Printing very basic mathematical functions

The easiest demonstration is to plot a simple exponential function

```python
import numpy as np

# Define our x range, -1 to 99
x = np.arange(-1, 100)

# Define our y function, exponential is x ** 2
y = x ** 2

print(f"First 10 elements of x => {x[:10]}")
print(f"First 10 elements of y => {y[:10]}")
```

```python
plt.figure(figsize=(10,10))
plt.plot(x, y)
plt.show()
```
